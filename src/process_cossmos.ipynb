```python
#@title function definition

import numpy as np
import pandas as pd

def get_prefix_before_number(s):
    in_quote = False
    for i, ch in enumerate(s):
        if ch == "'":
            in_quote = not in_quote
        elif ch.isdigit() and not in_quote:
            return s[:i]
    return s

def reorder_by_number(s):
    # Split into parts like ['1732-G', '1731-A', ...]
    parts = s.strip().split()
    # Sort based on the numeric part before the dash
    sorted_parts = sorted(parts, key=lambda x: int(x.split('-')[0]))
    # Join back into a string
    return ' '.join(sorted_parts)

def reorder_and_get_range(s):
    # Split into individual parts
    parts = s.strip().split()

    # 192-U 191-G 190-U 189L-G 189K-U

    # Extract number-token pairs
    number_token_pairs = [(int(remove_letters(p.split('-')[0])), p.split('-')[1], str(remove_letters(p.split('-')[0]))+'-'+p.split('-')[1]) for p in parts]
    # Sort by the number
    number_token_pairs.sort(key=lambda x: x[0])

    # Get sorted string
    sorted_seq_str = ''.join([p[1] for p in number_token_pairs])
    sorted_str = ' '.join([p[2] for p in number_token_pairs])

    # Get start and end numbers
    start_number = number_token_pairs[0][0]
    end_number = number_token_pairs[-1][0]

    return sorted_str, sorted_seq_str, start_number, end_number

def remove_letters(s): # 189L -> 189
    return ''.join(filter(str.isdigit, s))
```

```python
data_1x1 = pd.read_table('PlainText/CoSSMos_cryoEM_haripin3_April-20-2025.txt', sep='\t')
data_1x1.insert(2, 'Aseq_ChainID', np.nan)
data_1x1.insert(3, 'Bseq_ChainID', np.nan)
data_1x1.insert(4, 'Aseq_selection', np.nan)
data_1x1.insert(5, 'Bseq_selection', np.nan)
data_1x1.insert(6, 'Aseq_reorder', np.nan)
data_1x1.insert(7, 'Bseq_reorder', np.nan)
data_1x1.insert(8, 'Aseq_num_reorder', np.nan)
data_1x1.insert(9, 'Bseq_num_reorder', np.nan)
```

```python
for index in data_1x1.index:
  aseq_info = data_1x1.loc[index, 'A Residue Conformation 3']
  bseq_info = data_1x1.loc[index, 'B Residue Conformation 3']
  aseq = data_1x1.loc[index, 'Aseq']
  bseq = data_1x1.loc[index, 'Bseq']
  aseq_num = data_1x1.loc[index, 'Aseq_num']
  bseq_num = data_1x1.loc[index, 'Bseq_num']

  a_chain = get_prefix_before_number(aseq_info)
  b_chain = get_prefix_before_number(bseq_info)

  aseq_num_reorder, aseq_seq_reorder, a_start, a_end = reorder_and_get_range(aseq_num)
  bseq_num_reorder, bseq_seq_reorder, b_start, b_end = reorder_and_get_range(bseq_num)

  data_1x1.loc[index, 'Aseq_ChainID'] = a_chain
  data_1x1.loc[index, 'Bseq_ChainID'] = b_chain
  data_1x1.loc[index, 'Aseq_selection'] = f"{a_chain}:{a_start}-{a_end}"
  data_1x1.loc[index, 'Bseq_selection'] = f"{b_chain}:{b_start}-{b_end}"
  data_1x1.loc[index, 'Aseq_reorder'] = aseq_seq_reorder
  data_1x1.loc[index, 'Bseq_reorder'] = bseq_seq_reorder
  data_1x1.loc[index, 'Aseq_num_reorder'] = aseq_num_reorder
  data_1x1.loc[index, 'Bseq_num_reorder'] = bseq_num_reorder

```
